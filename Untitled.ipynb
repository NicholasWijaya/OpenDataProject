Display 2012-2018 drug-related deaths per county:

```python
filepath_2 = "D:\\CODE1161\\OpenDataProject\\countyData.csv"
county_data = pd.read_csv(filepath_2)
county_data.head()
```

Let's compare the deaths by County year-to-year:

```python
county_names = list(county_data.County.unique())

with plt.style.context("dark_background"):
    for name in county_names:
        # check with prescription rate and drug-related deaths
        total_death_annual = county_data[["Year", "Total Deaths"]][county_data["County"] == name].groupby("Year")
        plt.plot(total_death_annual,"x-")
plt.show()
```
